```python
import pandas as pd 
import io 
#Import towns into pandas dataframe 
df = pd.read_csv(io.BytesIO(uploaded['towns_copy.csv'])) 
print(df)
```

```python
import geopy
```

```python
#Using Nominatim API for geocoding
from geopy.geocoders import Nominatim
```

```python
#Assigning a username
geolocator = Nominatim(user_agent="My_geocoder")
```

```python
#Adding India to town lists.So that search can be more location specific
df["town_details"]=df["town"]+",India"
```

```python
#Geocoding
df["Coordinates"]=df["town_details"].apply(geolocator.geocode)
```

```python
#This line would give longitude of a place if coordinates are present.Otherwise it would be return None
df["Longitude"]=df["Coordinates"].apply(lambda x:x.longitude if x!=None else None)
```

```python
#This line would give latiude of a place if cordinates are present.Otherwise it would be return None
df["Latitude"]=df["Coordinates"].apply(lambda x:x.latitude if x!=None else None)
```
